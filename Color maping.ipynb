```python
def filtering(image,f=2):
    (rows,col,channels) = image.shape
    
    stride = 5
    
    n_rows = int(1+(rows-f)/stride)
    n_col = int(1+(col-f)/stride)
    n_channels = channels
    
    n_img = np.zeros((n_rows,n_col,n_channels))
    
    for h in range(n_rows):
        for w in range(n_col):
            for n in range(channels):
                vert_start = h*stride
                vert_end = vert_start + f
                hori_start = w*stride
                hori_end = hori_start + f
                
                n_slice = image[vert_start:vert_end,hori_start:hori_end,n]
                
                n_img[h,w,n] = np.mean(n_slice,dtype = int)
    return n_img
    
```

```python
a  = filtering(image, f= 1000)
print(a.shape)
plt.imshow(a)
plt.show()
```
