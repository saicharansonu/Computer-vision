# IMPORTING THE LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

# LOADING THE IMAGE

```python
img = plt.imread(r"C:/Users/G.VENKATARAMANA/DataSets/cell-images-for-detecting-malaria/malaria-bounding-boxes/train/010961af-b38c-49de-aca0-e3732d73d414.jpg")
```

# A KERNEL WITH A SIZE 20 X 20 

The kernel can be any size.

You can define based on the image size.

```python
filters = np.zeros((20,20))
```

# PLOTTING THE IMAGE

```python
plt.imshow(img)
plt.show()
```

# CONVERTING INTO 2D

```python
blc_wht = img.mean(axis = 2)
```

# PLOTTING THE 2D IMAGE

```python
plt.imshow(blc_wht,'gray')
plt.show()
```

# CALCULATING THE FILTER DISTANCE 

```python
for x in range(20):
    for y in range(20):
        dist = (x-10)**2 + (y-10)**2
        filters[x,y] = np.exp(-dist/50)
```

# APPLYING THE CONVOLUTION

```python
convolution = convolve2d(blc_wht,filters)
```

# PLOTTING THE CONVOLVED IMAGE

```python
plt.imshow(convolution,'gray')
plt.show()
```

# EDGE DETECTION

# KERNEL FOR DETECTING THE HORIZONTAL EDGES

```python
gx = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])
```

# KERNEL FOR DETECTING THE VERTICAL EDGES

```python
gy = gx.T
```

# PERFORMING THE HORIZONTAL EDGE FILTER TO THE IMAGE

```python
horitanzal_edges = convolve2d(blc_wht,gx)
```

Ploting the horizantally detected edges

```python
plt.imshow(horitanzal_edges,'gray_r')
plt.show()
```

# APPLYING THE VERTICAL EDGE FILTER TO THE IMAGE

```python
vertical_edges = convolve2d(blc_wht,gy)
```

Ploting the vertically detected edges

```python
plt.imshow(vertical_edges,'gray_r')
plt.show()
```

# CALCULATING THE KERNEL FOR DETECTING ALL THE EDGES

```python
g = np.sqrt(horitanzal_edges*horitanzal_edges + vertical_edges*vertical_edges)
```

# PLOTTING THE IMAGE WITH ALL THE DETECTED EDGES

```python
plt.imshow(g,cmap ='gray_r')
```
